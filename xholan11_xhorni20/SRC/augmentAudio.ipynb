```python
import torch
import torchaudio
import torchaudio.functional as F
from scipy.io import wavfile
from glob import glob
```

```python
# Apply effects
def apply_effect(waveform, sample_rate, effect):
    effector = torchaudio.io.AudioEffector(effect=effect)
    return effector.apply(waveform, sample_rate)
```

```python
from IPython.display import Audio

from torchaudio.utils import download_asset

SAMPLE_WAV = download_asset("tutorial-assets/steam-train-whistle-daniel_simon.wav")
SAMPLE_RIR = download_asset("tutorial-assets/Lab41-SRI-VOiCES-rm1-impulse-mc01-stu-clo-8000hz.wav")
SAMPLE_SPEECH = download_asset("tutorial-assets/Lab41-SRI-VOiCES-src-sp0307-ch127535-sg0042-8000hz.wav")
SAMPLE_NOISE = download_asset("tutorial-assets/Lab41-SRI-VOiCES-rm1-babb-mc01-stu-clo-8000hz.wav")
noise, _ = torchaudio.load(SAMPLE_NOISE)
```

```python
# dir_name='data/target_train/'
# dst = 'data/target_train_augment/'
dir_name='data/non_target_train/'
dst = 'data/non_target_train_augment/'

aug_idx=1

for f in glob(dir_name + '/*.wav'):
    print('Processing file: ', f)
    # Load the data
    waveform, sample_rate = torchaudio.load(f, channels_first=False)
    # sample_rate, s = wavfile.read(f)
    # print(s[10000])
    filename =f.split('/')[len(f.split('/'))-1].replace('.wav','')
    
    # Define effects
    effect = ",".join(
        [
            "lowpass=frequency=400:poles=1",  # apply single-pole lowpass filter
            "atempo=1.5",  # reduce the speed
            # Applying echo gives some dramatic feeling
        ],
    )

    
    new_wav = apply_effect(waveform, sample_rate, effect)
    # noise = noise[:, : waveform.shape[1]]

    # snr_dbs = torch.tensor([20, 10, 3])
    # noisy_speeches = F.add_noise(waveform, noise, snr_dbs)
    torchaudio.save(dst + filename + f'{aug_idx}.wav',new_wav,sample_rate,channels_first=False)
    
```
